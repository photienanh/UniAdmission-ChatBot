```python
from retriever import *
import os
```

```python
retriever = DataRetriever(
    search_config=WebsearchConfig(page_timeout=30, file_timeout=10),
    rag_config=RagConfig("intfloat/multilingual-e5-small"),
    splitter_config=SplitterConfig(chunk_overlap=0)
)
await retriever.start()
```

```python
os.environ["BRAVE_SEARCH_API_KEY"] = "BSAbUIq8YC6VrPhwp688ST6Vtz7cyrH"
os.environ["WEB_SEARCH_SSL"] = str(False)
```

```python
web_query = "ba công khai đại học bách khoa hà nội"
rag_query = web_query.lower()
web_sources = await retriever.websearch(
    web_query=web_query,
    k_pages=5,
    domain_restrict=False,
    engine="brave",
    include_pdf=True,
    include_image=False
)
rag_sources = await retriever.rag_retrieve(
    web_sources=web_sources,
    rerank_query=web_query,
    rag_query=web_query,
    k_docs=100,
    rerank_chunk=True,
    merge_table=True,
    merge_neighbor=False
)
import shutil, json
shutil.rmtree("z_output")
os.makedirs("z_output", exist_ok=True)
total = ""
for index, rag_source in enumerate(rag_sources):
    with open(f"z_output/{index}.json", 'w', encoding='utf-8') as file:
        prefix = f'[{rag_source["title"]}]({rag_source["url"]})'
        if "file_type" in rag_source:
            prefix += ":" + f'[{rag_source.get("file_title")}]({rag_source.get("file_url")})'
        total += "-"*50 + "\n" + prefix + "\n" + f"{rag_source["text"]}\n"
        file.write(json.dumps(rag_source, ensure_ascii=False))
with open(f"z_output/total.md", 'w', encoding='utf-8') as file:
    file.write(total)
formater = SourceFormat()
text = formater(rag_sources)
with open(f"z_output/formatted.md", 'w', encoding='utf-8') as file:
    file.write(text)
```
