```python
ws = Websearch(
    "intfloat/multilingual-e5-small",
    "Qwen/Qwen3-4B",
    "ms-marco-MultiBERT-L-12",
    "cpu",
    chunk_size=512,
    chunk_overlap=16
)
await ws.start()
```

```python
web_query = "Chương trình đào tạo nghành trí tuệ nhân tạo - Đại học công nghệ uet"
rag_query = web_query.lower()
web_sources, rag_sources = await ws(
    web_query,
    rag_query,
    5,
    5,
    False,
    "brave",
    True,
    False,
    True,
    True,
    0.5,
    neighbor_previous_k=0,
    neighbor_next_k=1
)
import shutil, json
shutil.rmtree("z_output")
os.makedirs("z_output", exist_ok=True)
total = ""
for index, rag_source in enumerate(rag_sources):
    with open(f"z_output/{index}.json", 'w', encoding='utf-8') as file:
        total += "-"*50 + "\n" + f"{rag_source["text"]}\n"
        file.write(json.dumps(rag_source, ensure_ascii=False))
with open(f"z_output/total.md", 'w', encoding='utf-8') as file:
    file.write(total)
```
