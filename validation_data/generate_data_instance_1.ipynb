```python
DOMAIN = "http://127.0.0.1:8000"

import requests
def login(username: str, password: str) -> str:
    url = f"{DOMAIN}/login"
    payload = {
        "username": username,
        "password": password
    }
    response = requests.post(url, json=payload)
    token = response.headers.get("set-cookie")
    token = str(token).split("jwt=")[-1].split(";")[0]
    return token
def chat(token: str, model_id: str, message: str) -> dict:
    payload = {
        "text": message,
        "model_id": model_id,
        "session_id": None,
        "params": {
            "engine_type": "brave",
            "domain_restrict": False,
            "k_docs": 5,
            "k_pages": 3,
            "max_tokens": 4096,
            "temperature": 0.7,
            "top_p": 0.9,
            "top_k": 16,
            "max_history": 0
        }
    }
    url = f"{DOMAIN}/chat"
    headers = {
        "jwt": token
    }
    response = requests.post(url, json=payload, cookies=headers)
    return response.json()
def get_result(url: str) -> str:
    response = requests.post(url)
    return response.text
```

```python
model_id = "Qwen/Qwen3-4B-LoRA-v1"

```

```python
import traceback
import time
with open("question_final.md", 'r', encoding='utf-8') as file:
    lines = file.readlines()
questions = []
for line in lines:
    line = line.strip()
    if len(line) > 0:
        questions.append(line.split(": ")[-1].strip())
for index in range(len(questions)):
    try:
        question = questions[index]
        print("-"*10 + "Question" + "-"*10)
        print(question)
        token = login("data2", "data2")
        time.sleep(1)
        data = chat(token, model_id, question)
        result_url = data["result_url"]
        time.sleep(1)
        result = get_result(result_url)
        # print("-"*10 + "Result" + "-"*10)
        # print(result)
        print(index)
        time.sleep(1)
    except:
        traceback.print_exc()
```
